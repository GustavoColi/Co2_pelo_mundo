```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')
import hvplot.pandas
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv")
```

```python
# dados em cache para melhorar o desempenho do painel
if 'data' not in pn.state.cache.keys():

    df = pd.read_csv('https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv')

    pn.state.cache['data'] = df.copy()

else: 

    df = pn.state.cache['data']
```

### (0) Algum pré-processamento de dados menor

```python
# Preencha NAs com 0s e crie a coluna PIB per capita
df = df.fillna(0)
df['gdp_per_capita'] = np.where(df['population']!= 0, df['gdp']/ df['population'], 0)
```

```python
# Tornar o pipeline do DataFrame interativo
idf = df.interactive()
```

### (1) Emissão de CO2 ao longo do tempo por continente

```python
# Definir widgets do painel
linha_temporal = pn.widgets.IntSlider(name='Linha Temporal', start=1750, end=2020, step=5, value=1850)
linha_temporal
```

```python
# Botões de rádio para medidas de CO2
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['co2', 'co2_per_capita',],
    button_type='success'
)
```

```python
continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_pipeline = (
    idf[
        (idf.year <= linha_temporal) &
        (idf.country.isin(continents))
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_pipeline
```

```python

co2_plot = co2_pipeline.hvplot(x = 'year', by='country', y=yaxis_co2,line_width=2, title="Emissão de CO2 por continente")
co2_plot
```

### (2) Tabela - Emissão de CO2 ao longo do tempo por continente

```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
co2_table
```

### (3) CO2 vs GDP scatterplot

```python
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == linha_temporal) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country', 'year', 'gdp_per_capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x='gdp_per_capita', 
                                                                y='co2', 
                                                                by='country', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=False, 
                                                                height=500, 
                                                                width=500)
co2_vs_gdp_scatterplot
```

### (4) Gráfico de barras com fontes de CO2 por continente

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['coal_co2', 'oil_co2', 'gas_co2'], 
    button_type='success'
)

continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_source_bar_pipeline = (
    idf[
        (idf.year == linha_temporal) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='country', 
                                                     y=yaxis_co2_source, 
                                                     title='Fonte de CO2 por continente')
co2_source_bar_plot
```

### Criando o Dashboard

```python
#Layout usando o Template
template = pn.template.FastListTemplate(
    title='Emissão de CO2 pelo Mundo - dashboard', 
    sidebar=[pn.pane.Markdown("# Emissões de CO2 e Mudanças Climáticas"), 
             pn.pane.Markdown("#### As emissões de dióxido de carbono são um dos principais motores de mudança climática em nosso planeta. Pesquisadores da área e a população no geral reconhecem que devemos tomar medidas preventivas e evitar os impactos da mudança climática ao longo prazo. O mundo precisa urgentemente controlar as emissões de CO2."), 
             pn.pane.PNG('climate_day.png', sizing_mode='scale_both'),
             pn.pane.Markdown("## Settings"),   
             linha_temporal],
    main=[pn.Row(pn.Column(yaxis_co2, 
                           co2_plot.panel(width=700), margin=(0,25)), 
                 co2_table.panel(width=500)), 
          pn.Row(pn.Column(co2_vs_gdp_scatterplot.panel(width=600), margin=(0,25)), 
                 pn.Column(yaxis_co2_source, co2_source_bar_plot.panel(width=600)))],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)
# template.show()
template.servable();
```
